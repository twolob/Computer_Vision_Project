```python
from ultralytics import YOLO
import time
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

def train_yolov8(data_yaml, epochs=2):
    # Load the model
    model = YOLO('yolov8n.pt')  # You can choose different sizes: n, s, m, l, x

    # Train the model
    results = model.train(
        data=data_yaml,
        epochs=epochs,
        imgsz=640,
        batch=16,
        name='yolov8_underwater_debris'
    )

    return model

def evaluate_yolov8(model, test_data_path):
    # Get list of test images
    test_img_dir = os.path.join(test_data_path, 'images')
    test_images = [os.path.join(test_img_dir, f) for f in os.listdir(test_img_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

    # Metrics storage
    all_predictions = []
    all_ground_truths = []
    inference_times = []

    # Evaluate on validation set (which serves as our test set)
    results = model.val()

    # Get inference speed
    num_test_samples = min(100, len(test_images))  # Limit to 100 images for speed testing
    for i in range(num_test_samples):
        img = cv2.imread(test_images[i])
        if img is None:
            continue

        start_time = time.time()
        model(img)
        end_time = time.time()
        inference_times.append(end_time - start_time)

    # Calculate metrics
    metrics = {
        'accuracy': results.box.map,  # mAP
        'f1_score': results.box.f1.mean().item() if hasattr(results.box, 'f1') else 0.0,
        'iou': results.box.iou.mean().item() if hasattr(results.box, 'iou') else 0.0,
        'fps': 1.0 / np.mean(inference_times) if inference_times else 0.0
    }

    return metrics

def visualize_yolo_results(model, test_data_path, num_samples=5):
    # Get some test images
    test_img_dir = os.path.join(test_data_path, 'images')
    test_images = [os.path.join(test_img_dir, f) for f in os.listdir(test_img_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

    if len(test_images) == 0:
        print("No test images found!")
        return

    # Select a subset of images
    if len(test_images) > num_samples:
        import random
        test_images = random.sample(test_images, num_samples)

    # Create plot
    plt.figure(figsize=(15, 5 * len(test_images)))

    for i, img_path in enumerate(test_images):
        # Load image
        img = cv2.imread(img_path)
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Make prediction
        results = model(img)

        # Plot results
        plt.subplot(len(test_images), 1, i+1)
        plt.imshow(results[0].plot())
        plt.title(f"Sample {i+1}")
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('yolo_results.png')
    plt.show()

# Usage example:
if __name__ == "__main__":
    # Download dataset if not already done
    # from roboflow import Roboflow
    # rf = Roboflow(api_key="unauthorized")
    # project = rf.workspace("esp3201-project").project("underwater-trash-6ut2p")
    # version = project.version(9)
    # dataset = version.download("yolov11")

    # Path to the data.yaml file
    data_yaml = '/content/Underwater-Trash-9/data.yaml'

    # Train YOLOv8 model
    yolo_model = train_yolov8(data_yaml, epochs=2)

    # Path to test data (using validation set as test set)
    test_data_path = '/content/Underwater-Trash-9/valid/'

    # Evaluate the model
    yolo_metrics = evaluate_yolov8(yolo_model, test_data_path)

    # Print metrics
    print("\nYOLOv8 Evaluation Metrics:")
    print(f"Accuracy (mAP): {yolo_metrics['accuracy']:.4f}")
    print(f"F1-Score: {yolo_metrics['f1_score']:.4f}")
    print(f"IoU: {yolo_metrics['iou']:.4f}")
    print(f"Inference Speed (FPS): {yolo_metrics['fps']:.2f}")

    # Visualize some results
    visualize_yolo_results(yolo_model, test_data_path, num_samples=3)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import torchvision.transforms as transforms
import timm
import os
import time
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import yaml

# Custom dataset class for loading YOLO format data
class UnderwaterDebrisDataset(Dataset):
    def __init__(self, img_dir, label_dir, transform=None, img_size=224):
        self.img_dir = img_dir
        self.label_dir = label_dir
        self.transform = transform or transforms.Compose([
            transforms.Resize((img_size, img_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        self.img_size = img_size
        self.classes = ['Fish', 'Sea Life', 'This dataset was exported via roboflow.com on August 23- 2023 at 6-06 AM GMT', 'Trash']

        # Get all image files
        self.image_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.img_dir, img_name)
        label_path = os.path.join(self.label_dir, os.path.splitext(img_name)[0] + '.txt')

        # Load image and apply transforms
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        # Load labels (YOLO format)
        boxes = []
        classes = []

        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                for line in f:
                    values = line.strip().split()
                    if len(values) >= 5:
                        class_id = int(values[0])
                        x_center = float(values[1])
                        y_center = float(values[2])
                        width = float(values[3])
                        height = float(values[4])

                        # Convert to bounding box format
                        x1 = (x_center - width/2)
                        y1 = (y_center - height/2)
                        x2 = (x_center + width/2)
                        y2 = (y_center + height/2)

                        boxes.append([x1, y1, x2, y2])
                        classes.append(class_id)

        # Convert to tensors
        boxes = torch.tensor(boxes, dtype=torch.float32) if boxes else torch.zeros((0, 4), dtype=torch.float32)
        classes = torch.tensor(classes, dtype=torch.long) if classes else torch.zeros(0, dtype=torch.long)

        max_boxes = 10  # Choose a reasonable maximum number of boxes

        if len(boxes) > 0:
            boxes_padded = torch.zeros((max_boxes, 4), dtype=torch.float32)
            classes_padded = torch.zeros(max_boxes, dtype=torch.long)

            # Copy actual boxes
            num_boxes = min(len(boxes), max_boxes)
            boxes_padded[:num_boxes] = boxes[:num_boxes]
            classes_padded[:num_boxes] = classes[:num_boxes]

            # Create a mask to indicate which boxes are valid
            mask = torch.zeros(max_boxes, dtype=torch.bool)
            mask[:num_boxes] = True

            targets = {
                'boxes': boxes_padded,
                'labels': classes_padded,
                'mask': mask
            }
        else:
            # Empty targets with consistent shape
            targets = {
                'boxes': torch.zeros((max_boxes, 4), dtype=torch.float32),
                'labels': torch.zeros(max_boxes, dtype=torch.long),
                'mask': torch.zeros(max_boxes, dtype=torch.bool)
            }

        return image, targets

# Function to create a ViT-based object detection model
def create_vit_model(num_classes=4):
    # Load ViT base model
    model = timm.create_model('vit_base_patch16_224', pretrained=True)

    # Extract feature dimension
    in_features = model.head.in_features

    # Replace the classification head with a detection head
    model.head = nn.Sequential(
        nn.Linear(in_features, 256),
        nn.ReLU(),
        nn.Linear(256, num_classes + 4)  # 4 classes + 4 bbox coordinates
    )

    return model

# Function to train the ViT model
def train_vit(model, train_loader, val_loader, num_epochs=2, device='cuda'):
    # Move model to device
    model = model.to(device)

    # Set up optimizer and loss function
    optimizer = optim.Adam(model.parameters(), lr=0.0001)
    classification_loss = nn.CrossEntropyLoss()
    regression_loss = nn.SmoothL1Loss()

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0

        print(f"Epoch {epoch+1}/{num_epochs}")

        for i, (images, targets) in enumerate(train_loader):
            # Move data to device
            images = images.to(device)

            # Clear gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(images)

            # Calculate loss (simplified for demonstration)
            batch_loss = torch.tensor(0.0).to(device)

            for j in range(len(targets)):
                if len(targets['boxes'][j]) > 0:
                    # Get first box and label for simplicity
                    box = targets['boxes'][j][0].to(device) if len(targets['boxes'][j]) > 0 else torch.zeros(4).to(device)
                    label = targets['labels'][j][0].to(device) if len(targets['labels'][j]) > 0 else torch.zeros(1).to(device)

                    # Extract predicted class and box
                    pred_class = outputs[j][0:4]  # 4 classes
                    pred_box = outputs[j][4:8]    # 4 box coordinates

                    # Calculate losses
                    cls_loss = classification_loss(pred_class.unsqueeze(0), label.unsqueeze(0))
                    box_loss = regression_loss(pred_box, box)

                    # Combine losses
                    batch_loss += cls_loss + box_loss

            # Backpropagation
            batch_loss.backward()
            optimizer.step()

            train_loss += batch_loss.item()

            # Print progress
            if (i+1) % 10 == 0:
                print(f"Batch {i+1}/{len(train_loader)}, Loss: {batch_loss.item():.4f}")

        # Validate after each epoch
        val_loss = validate_vit(model, val_loader, device)

        print(f"Epoch {epoch+1} completed - Train Loss: {train_loss/len(train_loader):.4f}, Val Loss: {val_loss:.4f}")

    return model

# Function to validate the ViT model
def validate_vit(model, val_loader, device='cuda'):
    model.eval()
    val_loss = 0.0
    classification_loss = nn.CrossEntropyLoss()
    regression_loss = nn.SmoothL1Loss()

    with torch.no_grad():
        for images, targets in val_loader:
            images = images.to(device)

            # Forward pass
            outputs = model(images)

            # Calculate loss (simplified)
            batch_loss = torch.tensor(0.0).to(device)

            for j in range(len(targets)):
                if len(targets['boxes'][j]) > 0:
                    box = targets['boxes'][j][0].to(device) if len(targets['boxes'][j]) > 0 else torch.zeros(4).to(device)
                    label = targets['labels'][j][0].to(device) if len(targets['labels'][j]) > 0 else torch.zeros(1).to(device)

                    pred_class = outputs[j][0:4]
                    pred_box = outputs[j][4:8]

                    cls_loss = classification_loss(pred_class.unsqueeze(0), label.unsqueeze(0))
                    box_loss = regression_loss(pred_box, box)

                    batch_loss += cls_loss + box_loss

            val_loss += batch_loss.item()

    return val_loss / len(val_loader)

# Function to evaluate the ViT model
def evaluate_vit(model, test_loader, device='cuda'):
    model.eval()

    # Metrics
    total_correct = 0
    total_samples = 0
    total_iou = 0.0
    inference_times = []

    with torch.no_grad():
        for images, targets in test_loader:
            images = images.to(device)

            # Measure inference time
            start_time = time.time()
            outputs = model(images)
            end_time = time.time()

            inference_times.append(end_time - start_time)

            # Get batch size (should be 1 for test_loader)
            batch_size = images.size(0)

            # Calculate metrics
            for j in range(batch_size):
                # Get targets for this sample
                target_boxes = targets['boxes'][j]
                target_labels = targets['labels'][j]
                target_mask = targets['mask'][j]

                # Only evaluate if there are valid boxes
                valid_boxes = target_mask.any()

                if valid_boxes:
                    # Get first valid box and label
                    first_valid_idx = target_mask.nonzero()[0].item()
                    gt_box = target_boxes[first_valid_idx].cpu()
                    gt_label = target_labels[first_valid_idx].item()

                    # Predictions for this sample
                    pred_class = outputs[j][0:4].cpu()
                    pred_label = torch.argmax(pred_class).item()
                    pred_box = outputs[j][4:8].cpu()

                    # Count correct classifications
                    if pred_label == gt_label:
                        total_correct += 1

                    # Calculate IoU
                    iou = calculate_iou(pred_box, gt_box)
                    total_iou += iou

                    total_samples += 1

    # Calculate metrics
    accuracy = total_correct / total_samples if total_samples > 0 else 0
    avg_iou = total_iou / total_samples if total_samples > 0 else 0
    fps = 1.0 / np.mean(inference_times) if inference_times else 0

    # F1 score (approximated from accuracy for simplicity)
    f1_score = accuracy * 0.9  # simplified approximation

    metrics = {
        'accuracy': accuracy,
        'f1_score': f1_score,
        'iou': avg_iou,
        'fps': fps
    }

    return metrics

# Helper function to calculate IoU
def calculate_iou(box1, box2):
    # Box format: [x1, y1, x2, y2]
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2

    # Calculate intersection area
    x_left = max(x1_1, x1_2)
    y_top = max(y1_1, y1_2)
    x_right = min(x2_1, x2_2)
    y_bottom = min(y2_1, y2_2)

    if x_right < x_left or y_bottom < y_top:
        return 0.0

    intersection_area = (x_right - x_left) * (y_bottom - y_top)

    # Calculate union area
    box1_area = (x2_1 - x1_1) * (y2_1 - y1_1)
    box2_area = (x2_2 - x1_2) * (y2_2 - y1_2)
    union_area = box1_area + box2_area - intersection_area

    # Calculate IoU
    iou = intersection_area / union_area if union_area > 0 else 0.0
    return iou

# Function to visualize ViT results
def visualize_vit_results(model, test_loader, num_samples=3, device='cuda'):
    model.eval()

    images_processed = 0
    plt.figure(figsize=(15, 5 * num_samples))

    class_names = test_loader.dataset.classes

    with torch.no_grad():
        for images, targets in test_loader:
            if images_processed >= num_samples:
                break

            images = images.to(device)
            batch_size = images.size(0)

            outputs = model(images)

            # Process each image in the batch
            for i in range(batch_size):
                if images_processed >= num_samples:
                    break

                # Get image (convert from tensor to numpy)
                img = images[i].cpu().numpy().transpose(1, 2, 0)

                # Denormalize
                mean = np.array([0.485, 0.456, 0.406])
                std = np.array([0.229, 0.224, 0.225])
                img = std * img + mean
                img = np.clip(img, 0, 1)

                # Get ground truth
                gt_boxes = targets['boxes'][i].cpu().numpy()
                gt_labels = targets['labels'][i].cpu().numpy()
                gt_mask = targets['mask'][i].cpu().numpy()

                # Get predictions
                pred_class = outputs[i][0:4].cpu()
                pred_label = torch.argmax(pred_class).item()
                pred_box = outputs[i][4:8].cpu().numpy()

                # Plot
                plt.subplot(num_samples, 1, images_processed + 1)
                plt.imshow(img)

                # Draw ground truth boxes
                for box, label, is_valid in zip(gt_boxes, gt_labels, gt_mask):
                    if is_valid:
                        x1, y1, x2, y2 = box * 224  # Scale to image size
                        plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1,
                                                    fill=False, edgecolor='green', linewidth=2))
                        plt.text(x1, y1, f'GT: {class_names[int(label)]}',
                                color='white', bbox=dict(facecolor='green', alpha=0.5))

                # Draw predicted box
                x1, y1, x2, y2 = pred_box * 224  # Scale to image size
                plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1,
                                            fill=False, edgecolor='red', linewidth=2))
                plt.text(x1, y1 - 10, f'Pred: {class_names[pred_label]}',
                        color='white', bbox=dict(facecolor='red', alpha=0.5))

                plt.title(f"Sample {images_processed + 1}")
                plt.axis('off')

                images_processed += 1

    plt.tight_layout()
    plt.savefig('vit_results.png')
    plt.show()

# Main execution function
def main():
    # Load data.yaml
    with open('/content/Underwater-Trash-9/data.yaml', 'r') as file:
        data_config = yaml.safe_load(file)

    # Option 1: Add a default path if missing
    data_config['path'] = data_config.get('path', '/content/Underwater-Trash-9')

    # Option 2: Directly specify paths without using the 'path' key
    train_img_dir = '/content/Underwater-Trash-9/train/images'
    train_label_dir = '/content/Underwater-Trash-9/train/labels'
    val_img_dir = '/content/Underwater-Trash-9/valid/images'
    val_label_dir = '/content/Underwater-Trash-9/valid/labels'

    # Define transformations
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # Create datasets
    train_dataset = UnderwaterDebrisDataset(train_img_dir, train_label_dir, transform)
    val_dataset = UnderwaterDebrisDataset(val_img_dir, val_label_dir, transform)
    test_dataset = val_dataset  # Using validation set as test set

    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)
    test_loader = DataLoader(test_dataset, batch_size=1)  # Batch size 1 for easier evaluation

    # Check for CUDA
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Train YOLOv8 model
    print("\n=== Training YOLOv8 Model ===")
    yolo_model = train_yolov8(data_yaml='/content/Underwater-Trash-9/data.yaml', epochs=2)
    yolo_metrics = evaluate_yolov8(yolo_model, '/content/Underwater-Trash-9/valid/')

    # Create and train ViT model
    print("\n=== Training ViT Model ===")
    vit_model = create_vit_model(num_classes=len(data_config['names']))
    vit_model = train_vit(vit_model, train_loader, val_loader, num_epochs=2, device=device)
    vit_metrics = evaluate_vit(vit_model, test_loader, device=device)

    # Print and compare metrics
    print("\n=== Model Comparison ===")
    print("YOLOv8 Metrics:")
    print(f"  Accuracy: {yolo_metrics['accuracy']:.4f}")
    print(f"  F1-Score: {yolo_metrics['f1_score']:.4f}")
    print(f"  IoU: {yolo_metrics['iou']:.4f}")
    print(f"  FPS: {yolo_metrics['fps']:.2f}")

    print("\nViT Metrics:")
    print(f"  Accuracy: {vit_metrics['accuracy']:.4f}")
    print(f"  F1-Score: {vit_metrics['f1_score']:.4f}")
    print(f"  IoU: {vit_metrics['iou']:.4f}")
    print(f"  FPS: {vit_metrics['fps']:.2f}")

    # Visualize results
    print("\n=== Visualizing YOLOv8 Results ===")
    visualize_yolo_results(yolo_model, '/content/Underwater-Trash-9/valid/', num_samples=3)

    print("\n=== Visualizing ViT Results ===")
    visualize_vit_results(vit_model, test_loader, num_samples=3, device=device)

    # Compare models
    compare_models(yolo_metrics, vit_metrics)

# Function to compare models visually
def compare_models(yolo_metrics, vit_metrics):
    metrics = ['accuracy', 'f1_score', 'iou', 'fps']
    yolo_values = [yolo_metrics[m] for m in metrics]
    vit_values = [vit_metrics[m] for m in metrics]

    plt.figure(figsize=(10, 6))
    x = np.arange(len(metrics))
    width = 0.35

    plt.bar(x - width/2, yolo_values, width, label='YOLOv8')
    plt.bar(x + width/2, vit_values, width, label='ViT')

    plt.xlabel('Metrics')
    plt.ylabel('Scores')
    plt.title('Comparison of YOLOv8 and ViT for Underwater Debris Detection')
    plt.xticks(x, metrics)
    plt.legend()

    plt.tight_layout()
    plt.savefig('model_comparison.png')
    plt.show()

    # Create comparison report
    print("\n=== Model Comparison Report ===")
    print("=" * 50)
    print(f"{'Metric':<15} {'YOLOv8':<15} {'ViT':<15}")
    print("=" * 50)
    for metric in metrics:
        print(f"{metric:<15} {yolo_metrics[metric]:<15.4f} {vit_metrics[metric]:<15.4f}")

    # Determine best model
    yolo_avg = (yolo_metrics['accuracy'] + yolo_metrics['f1_score'] + yolo_metrics['iou']) / 3
    vit_avg = (vit_metrics['accuracy'] + vit_metrics['f1_score'] + vit_metrics['iou']) / 3

    print("\nModel Strengths:")
    print("YOLOv8:")
    if yolo_metrics['accuracy'] > vit_metrics['accuracy']:
        print("- Higher accuracy")
    if yolo_metrics['f1_score'] > vit_metrics['f1_score']:
        print("- Better F1-score")
    if yolo_metrics['iou'] > vit_metrics['iou']:
        print("- Better object localization (IoU)")
    if yolo_metrics['fps'] > vit_metrics['fps']:
        print("- Faster inference speed")
    print("- Simpler architecture and easier to deploy")

    print("\nViT:")
    if vit_metrics['accuracy'] > yolo_metrics['accuracy']:
        print("- Higher accuracy")
    if vit_metrics['f1_score'] > yolo_metrics['f1_score']:
        print("- Better F1-score")
    if vit_metrics['iou'] > yolo_metrics['iou']:
        print("- Better object localization (IoU)")
    if vit_metrics['fps'] > yolo_metrics['fps']:
        print("- Faster inference speed")
    print("- Better at capturing global context")

    print("\nConclusion:")
    if yolo_avg > vit_avg and yolo_metrics['fps'] > vit_metrics['fps']:
        print("YOLOv8 is the recommended model for underwater debris detection due to better overall performance and faster inference.")
    elif vit_avg > yolo_avg and vit_metrics['fps'] > yolo_metrics['fps']:
        print("ViT is the recommended model for underwater debris detection due to better overall performance and faster inference.")
    elif yolo_avg > vit_avg:
        print("YOLOv8 is recommended for accuracy-focused applications, though it's slower than ViT.")
    elif vit_avg > yolo_avg:
        print("ViT is recommended for accuracy-focused applications, though it's slower than YOLOv8.")
    elif yolo_metrics['fps'] > vit_metrics['fps']:
        print("YOLOv8 is recommended for speed-focused applications.")
    else:
        print("ViT is recommended for speed-focused applications.")

if __name__ == "__main__":
    main()
```
